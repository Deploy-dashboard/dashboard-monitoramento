```python
import requests

res = requests.get(f'http://10.0.10.22:41112/gw/reports/documentos/filter/CAED7029-2307:2025/null/null')
data = res.json()
display(data)
```

```python
dados = data[0]["dados"]
programas = {item["key"]: item["value"] for item in dados}

display(programas)
```

```python
id = programas.keys()
print(id)

sp = {}
for key in id:
    res_sp = requests.get(f'http://10.0.10.22:41112/gw/reports/documentos/filterssrs/CAED7029-2307:2025/58/{key}/ID_FONTE_DADO={key}&CD_PROGRAMA=undefined&DC_SOLICITACAO=undefined')
    data_sp = res_sp.json()
    sp[key] = data_sp[0]["dados"]

subprogramas = {
    k: [item["value"] for item in v]
    for k, v in sp.items()
}

display(subprogramas)
```
